```python
import numpy as np
import matplotlib.pyplot as plt

objects = ('H(x)', 'H(y)', 'H(y|x)', 'H(x|y)', 'H(x,y)', 'I(x,y)')
values = [0.637,0.637,0.462,0.462,1.099,0.175]

f=plt.figure()
plt.bar(x = objects, height = values, align='center', alpha=0.5)
plt.xticks(objects)
plt.ylabel('Values')
plt.bar(objects, values)
for a,b in zip(objects, values):
    plt.text(a, b, str(b))
plt.show()
f.savefig("Relationship.pdf", bbox_inches='tight')
```
